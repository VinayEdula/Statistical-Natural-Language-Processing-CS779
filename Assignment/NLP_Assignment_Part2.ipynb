## 1. Introduction to TF-IDF

**Term Frequency-Inverse Document Frequency (TF-IDF)** is a technique used to convert text documents into numerical feature vectors, capturing the importance of words within documents relative to a corpus (https://en.wikipedia.org/wiki/Tf%E2%80%93idf).

### Key Components:

- **Term Frequency (TF):** Measures how often a term appears in a document.
  $$
  TF(t, d) = \frac{\text{Number of times term } t \text{ appears in document } d}{\text{Total number of terms in document } d}
  $$

- **Inverse Document Frequency (IDF):** Measures how important a term is by considering the rarity of the term across all documents.
  $$
  IDF(t, D) = \log \left(\frac{\text{Total number of documents in the corpus}}{\text{Number of documents containing the term } t}\right)
  $$

- **TF-IDF Calculation:** Combines TF and IDF values to produce a weighted score for each term in a document.
  $$
  TF-IDF(t, d, D) = TF(t, d) \times IDF(t, D)
  $$

## 2. Introduction to K-Nearest Neighbors (KNN)

**K-Nearest Neighbors (KNN)** is a simple, non-parametric algorithm used for classification and regression. It predicts the class of a data point based on the classes of its nearest neighbors (https://en.wikipedia.org/wiki/K-nearest_neighbors_algorithm).

### Steps in KNN Classification:

1. **Distance Calculation:** Compute the distance between a test document and all training documents. The Euclidean distance is commonly used:
   $$
   d(x, y) = \sqrt{\sum_{i=1}^{n} (x_i - y_i)^2}
   $$

2. **Identify Neighbors:** Find the `k` closest training documents to the test document.

3. **Prediction:** Assign the most common class label among the `k` nearest neighbors to the test document.

### **TF-IDF Vectorization:**

```python
import nltk
# Download necessary NLTK resources
nltk.download('punkt')
```

**Approach:**




1. Initially the dataframe is divided into chunks of 1000 articles at a time, this was implemented using generators in python.
2. Each chunk is then sent to get_tf() function, it will return the TF of all the articles present in that chunk. Next we will update the tf_dictionary which is global tf dictionary which contains TF of all the chunks.

*tf_dictionary:* key is article name and value is a dictionary which contains tokens as key and tf of that token as value.

*get_tf(raw_articles):* Takes chunk of dataframe.Extracts text from each article, tokenizes it using NLTK.(Tried using spacy at first for tokenization but NLTK is extracting faster).Finds tf dictionary of all these documents one-by-one using formula (word: count/total_count) and returns it.

3. Once finding TF is completed. We will calculate the document frequency using compute_df(tf_dictionary) function.

*compute_df(tf_dictionary):* This function takes tf_dictionary as input and for each article, it will finds all unique tokens present in that article(which are keys of tf_dictionary itself) and finds the number of documents a token occurs and updates in the df_count dictionary and returns it.

4.  Next we will find tf-idf values. get_tf_idf(tf_dictionary, df_count) takes tf_dictionary and df_count dictionary. This function returns a *tf_idf_dictionary* which has keys as article name and values as tf-idf of entire article as a dictionary(calculated using the tf_dictionary and df_count, and applying the given formula in question) . Additionally i have stored category, name of article in this dictionary which will be used in next questions.

```python
import pandas as pd
import math
from collections import Counter
from nltk.tokenize import word_tokenize

all_tokens = set()


def get_tf(raw_articles):
    tf_dictionary = {}
    for rownum, article in raw_articles.iterrows():
        text = article['text']
        tokens = word_tokenize(text)  # Tried using spacy tokkenizer , SYSTEM ram is
        all_tokens.update(tokens)
        total_terms_in_document = len(tokens)
        term_count = Counter(tokens)
        document_tf = {term: count / total_terms_in_document for term, count in term_count.items()}
        document_tf['name'] = article['article']
        document_tf['category'] = article['category']
        tf_dictionary[article['article']] = document_tf
    return tf_dictionary


def compute_df(tf_dictionary):
    df_count = Counter()
    for tf_values in tf_dictionary.values():
        for term in tf_values:
            if term not in ['name', 'category']:  # Excludes non-token fields
                df_count[term] += 1
    return df_count

def get_tf_idf(tf_dictionary, df_count):
    total_documents = len(tf_dictionary)
    idf_dictionary = {term: math.log(total_documents / (df_count[term] + 1)) for term in df_count}  # Added +1 to avoid division by zero
    tf_idf_dictionary = {}
    for article_name, tf_values in tf_dictionary.items():
        tf_idf_values = {term: tf * idf_dictionary.get(term, 0) for term, tf in tf_values.items() if term not in ['name', 'category']}
        tf_idf_values['name'] = tf_values['name']
        tf_idf_values['category'] = tf_values['category']
        tf_idf_dictionary[article_name] = tf_idf_values
    return tf_idf_dictionary

def chunk_dataframe(df, chunk_size):
    num_rows = len(df)
    for start in range(0, num_rows, chunk_size):
        end = min(start + chunk_size, num_rows)
        yield df.iloc[start:end]
chunk_size = 1000
tf_dictionary = {}
for i, chunk in enumerate(chunk_dataframe(train_df, chunk_size)):
    tf = get_tf(chunk)
    tf_dictionary.update(tf)
    print(f"{(i + 1)*100*chunk_size / len(train_df) }% data processed for TF.")

df_count = compute_df(tf_dictionary)
tf_idf_dictionary = get_tf_idf(tf_dictionary, df_count)
print("TF-IDF computation completed.")
```

### **KNN Implementation:**

**Approach:**


***classify_document_KNN(test_text, tf_idf_dictionary, df_count, num_of_neighbours)***  is the function that finds the categories and names of K-Nearest articles and it simultaneously implements for 3 different distance metrices separately.

1. First it finds the tf-idf of test_vector using compute_tf_idf_vector_test_data method.


> *compute_tf_idf_vector_test_data(text, df_count):*  Takes text content of a test article and gives tf_idf_values of the corresponding article as a dictionary, It contains token as key and value as corresponding tf_idf_value. First we tokenize the text, then we will find the tf_values using Counter() in python which returns frequency of each token. Then we will find idf usinf df_count which found in earlier step.


2. Finds distance between the test vector to all the train vectors using compute_distance function. We will store the distances between train and test vector in 3 separate lists each per metric.

> *compute_distance(vec1, vec2):* Takes tf_idf dictionaries of two articles and returns three type of distances between them i.e cosine, euclidean, manhattan as a list(Asked in next questions so i implemented here). First we will take union of tokens of two articles and then we create actual tf-idf vectors of two articles from the two dictionaries, if the tf-idf values are not present in the dictionary for a particular token then we use 0 in that place.


3. Sort the distance lists as we need to find nearest neighbours in terms of distance.

4.  Finally we will get extract the neighbours details like categories using get_result_from_distances() function. To store the results separately given by 3 distance metrices we will use a dictionary named fnal_results, where key is distance metric name and value is corresponding results of that metric.

> get_result_from_distances(distances, num_of_neighbours): Takes distances and num_of_neighbours(a list) which contains multiple values. This function finds the k nearest neighbours details for each k present in the num_of_neighbours list.


```python
import numpy as np
import math
from collections import Counter
from nltk.tokenize import word_tokenize

def classify_document_KNN(test_text, tf_idf_dictionary, df_count, num_of_neighbours):
    test_vector = compute_tf_idf_vector_test_data(test_text, df_count)

    distances_euclidean = []
    distances_manhattan = []
    distances_cosine = []

    for name, values in tf_idf_dictionary.items():
        train_vector = values
        distance = compute_distance(test_vector, train_vector)
        distances_cosine.append((distance[0], values['name'], values['category']))
        distances_manhattan.append((distance[1], values['name'], values['category']))
        distances_euclidean.append((distance[2], values['name'], values['category']))

    distances_cosine.sort(key=lambda x: -x[0])
    distances_manhattan.sort(key=lambda x: x[0])
    distances_euclidean.sort(key=lambda x: x[0])

    final_result = {}
    final_result["euclidean"] = get_result_from_distances(distances_euclidean, num_of_neighbours)
    final_result["cosine"] = get_result_from_distances(distances_cosine, num_of_neighbours)
    final_result["manhattan"] = get_result_from_distances(distances_manhattan, num_of_neighbours)

    return final_result


def compute_tf_idf_vector_test_data(text, df_count):
    tokens = word_tokenize(text)
    term_count = Counter(tokens)
    total_terms_in_document = len(tokens)
    tf_values = {term: count / total_terms_in_document for term, count in term_count.items()}
    idf_values = {term: math.log(len(df_count) / (df_count[term] + 1)) for term in term_count}
    tf_idf_values = {term: tf_values.get(term, 0) * idf_values.get(term, 0) for term in tf_values}
    return tf_idf_values

def compute_distance(vec1, vec2):
    all_terms = set(vec1.keys()).union(set(vec2.keys()))
    all_terms.discard('name')
    all_terms.discard('category')
    vec1_extended = np.array([vec1.get(term, 0) for term in all_terms], dtype=np.float64)
    vec2_extended = np.array([vec2.get(term, 0) for term in all_terms], dtype=np.float64)

    distances = []
    norm1 = np.linalg.norm(vec1_extended)
    norm2 = np.linalg.norm(vec2_extended)
    if norm1 == 0 or norm2 == 0:
      distances.append(-1)
    else:
      distances.append(np.dot(vec1_extended, vec2_extended) / (norm1 * norm2)) #Cosine similarity
    distances.append(np.sum(np.abs(vec1_extended - vec2_extended)))  #Manhattan distance
    distances.append(np.sqrt(np.sum((vec1_extended - vec2_extended)**2))) #Euclidean distance
    return distances


def get_result_from_distances(distances, num_of_neighbours):
  temp_results=[]
  for num in num_of_neighbours:
    nearest_categories = Counter([category for dist, name, category in distances[:num]])
    nearest_names_cat_pair = [(name,category) for dist, name, category in distances[:num]]
    temp_results.append([num, nearest_categories,nearest_names_cat_pair])
  return temp_results
```

**Explanation of Testing Phase:**

1. test_KNN() takes num_of_neighbours( a list containing different K values to test) and number of test examples to test from test dataset.

2. It will iterate over test dataset and for each test article using classify_document_KNN() function, it will find the K Nearest neighbour details, From those neighbour details we will find the most common category in them and store the corresponding result. Simultaneously we will store the true categories in true_labels list.

3. Results was stored in a dictionary called prediction_labels, where key is distance metric name and value is again a dictionary. In value dictionary, keys are num_of_neighbours and values are list of predictions.

4. Next will evaluate the predictions using accuracy, recall, precision and f1 score. For all 3 distance metrices and all K values given for the function.

```python
#Testing

from collections import Counter,defaultdict
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score


def test_KNN(num_of_neighbours, no_of_test_examples):

    prediction_labels = {}
    prediction_labels["euclidean"] = defaultdict(list)
    prediction_labels["cosine"] = defaultdict(list)
    prediction_labels["manhattan"] = defaultdict(list)

    true_labels = []

    for index, article in test_df.iterrows():
        if no_of_test_examples == 0:
            break
        result = classify_document_KNN(article['text'], tf_idf_dictionary, df_count,num_of_neighbours)
        true_category = article['category']
        true_labels.append(true_category)
        for metric in result: #result is a dictionary with keys as euclidean, cosine, manhattan
          results = result[metric]
          for res in results: #Contains result for every k
            k = res[0]
            nearest_categories = res[1]
            prediction = nearest_categories.most_common(1)[0][0]
            prediction_labels[metric][k].append(prediction)
        no_of_test_examples -= 1

    final_result = {}
    final_result["euclidean"] = defaultdict(list)
    final_result["cosine"] = defaultdict(list)
    final_result["manhattan"] = defaultdict(list)

    for metric in prediction_labels:
      print(f"Distance metric choosen: {metric}")
      print("\n")
      pred_labels = prediction_labels[metric]
      for neighbour_size in pred_labels:
        ytrue = true_labels
        ypred = pred_labels[neighbour_size]
        accuracy = accuracy_score(ytrue, ypred)
        precision = precision_score(ytrue, ypred, average='weighted', zero_division=0)
        recall = recall_score(ytrue, ypred, average='weighted', zero_division=0)
        f1 = f1_score(ytrue, ypred,average='weighted')
        print(f"Choosen K={neighbour_size}")
        print(f"Accuracy: {accuracy}")
        print(f"Precision: {precision}")
        print(f"Recall: {recall}")
        print(f"F1 Score: {f1}")
        print("\n")
        final_result[metric][neighbour_size].append([accuracy,precision,recall,f1])
    return final_result
```

### **Optimizing k-NN for Large Datasets**

**KD Tree:**

1. KD tree is a binary tree where each node acts as a splitting point which divides the data points  into two parts.

2. One way of implementing KD-tree is recursively split the data based on median of one of dimension of data points. (https://www.baeldung.com/cs/k-d-trees)

3. This helps in reducing the search space of K-Nearest Neighbours. Because we will traverse the tree and reach leaf nodes, now ideally we will have very less data points compared to initial data points.

4. Finally the amount of calculations will be reduced compared to usual KNN model.

Creating TF-IDF data of 1000 points to give demo of KD-Trees and Ball-Trees(My earlier tf-idf values are stored in dictionary format, as i am using libraries for this part, these libraries expect it in array format).

```python

tokens = set()
for i, article in enumerate(tf_idf_dictionary):
  if i>1000:
    break
  article_dict = tf_idf_dictionary[article]
  tokens.update(article_dict.keys())

tokens = list(tokens)
data = []
labels = []
for i, article in enumerate(tf_idf_dictionary):
  if i>1000:
    break
  tfidf = []
  article_dict = tf_idf_dictionary[article]
  for token in tokens:
    if token!='name' and token!='category':
      value = article_dict.get(token, 0)
      tfidf.append(value)
  data.append(tfidf)
  labels.append(article_dict['category'])
```

Used KDTree function in sklearn library build the model on our data. Leaf_size contains the number of data points it need to store in leafs of tree, it will stop the construction of tree once it reached that leaf_size. During prediction it will apply usual KNN algorithm at leaf level.

```python
from sklearn.neighbors import KDTree
import numpy as np
knn_using_kd_tree = KDTree(data, leaf_size=5)
```

3. Using these data we can find the most common category in the neighbours.

**Ball Tree:**

1. Data is grouped into several balls, where each ball contains nearest data points interms of distance decided by certain radius.

2. During testing we can avoid many computations by ignoring the balls which are very distance to test points(by using centroid of data) and concentrating on balls which possibly contain our predictions.

### 1. Introduction to K-Means Clustering

K-Means is a popular clustering algorithm used to partition data into k distinct clusters. Each cluster is represented by its centroid, which is the mean of all points assigned to that cluster.

**Steps in K-Means Clustering:**
- **Initialization:** Choose a value of k (this corresponds to number of clusters). This is a hyper-parameter that you need to experiment with.
- **Assignment:** Assign each data point to the nearest centroid.
- **Update:** Recalculate centroids as the mean of the points assigned to each cluster.
- **Iteration:** Repeat the assignment and update steps until centroids no longer change significantly.

**Creating sparse data matrix to train KMeans:**

This part of code creates a sparse matrix of TF-IDF values. Creating 2d arrays of tf-idf vectors is inefficient because it takes lot of space and many entries are zeros in it, and training these arrays is computationally intensive. KMeans model in sklearn accepts sparse matrix and we only store non zero values in sparse matrix.

1. Each row in sparse matrix belongs to one particular article in dataset.

2. Each column in sparse matrix belongs to one unique token.

3. index [row, col] contains the tf-idf value of that entry.

```python
from scipy.sparse import csr_matrix
tokens_list = list(set(all_tokens))
tokens_list.remove('name')
tokens_list.remove('category')

tokens_list_index = {}
for ind, token in enumerate(tokens_list):
  tokens_list_index[token] = ind

row_indices = []
col_indices = []
data = []
article_names = []
labels = []
num_of_train_examples = 5000
#As most number of elements in the tf-idf vector are zeros, i am using sparse matrix
for i, article in enumerate(tf_idf_dictionary):
  if i==num_of_train_examples:
    break
  article_dict = tf_idf_dictionary[article]
  for token in tokens_list:
    value = article_dict.get(token, 0)
    if value > 0:  #Storing only non zero values in sparse matrix
      row_indices.append(i)
      col_indices.append(tokens_list_index[token])
      data.append(value)
  article_names.append(article)
  labels.append(article_dict['category'])

# Create a sparse matrix
sparse_matrix = csr_matrix((data, (row_indices, col_indices)), shape=(num_of_train_examples, len(tokens_list)))
```

**Explanation:**

*1. create_kmeans_model(data, k):*  Creates KMeans model using sklearn library and by taking data and K value and returns the corresponding centers.



*2. Performance_Evaluation_via_Metrics(ytrue, ypred):*  Returns accuracy_score, precision_score, recall_score, f1_score between ypred and ytrue.


*3. guess_center_to_label(sparse_matrix, center):* This method is implemented for the purpose of labelling the clusters. It will find the closest articles to the given cluster center and returns the details of those articles. While labelling the clusters we can see the type of articles and can assign the corresponding category.


*4. classify_article(article_dict,kmeans_model,tokens_list_index):* Takes a tfidf dictionary of a article and classifies the cluster center it belongs to. As the model does not accepts the dictionary directly we will convert it to CSR matrix form then predicts the cluster.

*5. testKMeans(kmeans_model,center_to_label,sparse_matrix, no_of_test_examples):*  Using all the above implemented methods and after manually labelling the clusters(center_to_label is dictionary which is labelled manually, given a index of cluster it will give corresponfing category), We can do testing using unseen samples. This model returns the true labels and predicted labels.



```python
from scipy.sparse import csr_matrix
from sklearn.cluster import KMeans
import numpy as np
from collections import Counter

def create_kmeans_model(data, k):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(data)
    centers = kmeans.cluster_centers_
    print("Cluster centers:", centers)
    return centers,kmeans

def guess_center_to_label(sparse_matrix, center):
    distances = []
    for i in range(sparse_matrix.shape[0]):
        row = sparse_matrix.getrow(i).toarray()
        distance = np.linalg.norm(row - center)
        distances.append((i, distance))
    closest_indices = [idx for idx, dist in sorted(distances, key=lambda x: x[1])[:10]]
    closest_labels = [labels[i] for i in closest_indices]
    closest_article_names = [article_names[i] for i in closest_indices]
    return closest_article_names,closest_labels,closest_indices


def classify_article(article_dict,kmeans_model,tokens_list_index):
    row_indices = []
    col_indices = []
    data = []
    for token, token_index in tokens_list_index.items():
        value = article_dict.get(token, 0)
        if value > 0:
            row_indices.append(0)
            col_indices.append(token_index)
            data.append(value)
    sparse_article_matrix = csr_matrix((data, (row_indices, col_indices)), shape=(1, len(tokens_list_index)))
    predicted_cluster = kmeans_model.predict(sparse_article_matrix)
    return predicted_cluster[0]

def testKMeans(kmeans_model,center_to_label,sparse_matrix, no_of_test_examples):
    ypred = []
    ytrue = []
    start=5001
    for i, article in enumerate(tf_idf_dictionary, start=5001):
        if i == no_of_test_examples+start:
            break
        article_dict = tf_idf_dictionary[article]
        true_category = article_dict['category']
        ytrue.append(true_category)
        closest_center_index = classify_article(article_dict,kmeans_model,tokens_list_index)
        predicted_label = center_to_label[closest_center_index]
        ypred.append(predicted_label)
    return ypred, ytrue

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
def Performance_Evaluation_via_Metrics(ytrue, ypred):
  accuracy = accuracy_score(ytrue, ypred)
  precision= precision_score(ytrue, ypred, average='weighted', zero_division=0)
  recall= recall_score(ytrue, ypred, average='weighted', zero_division=0)
  f1_scores = f1_score(ytrue, ypred,average='weighted')
  return accuracy, precision, recall, f1_scores
```

**1. How do KNN and KMeans differ fundamentally in their approach to classifying text data?**

**Methodology:**
KNN finds K nearest neighbours using distance/similarity measures and to find the output it uses techniques like majority of the nearest neighbours or weighted average of neighbours results.

Kmeans forms clusters in the data, it groups similar documents and it represents the entire group using centroid. So during testing we need to just find the distance from test vector to centroids and classify the test vector as the group of closest centroid.

**Cost:**
In KNN, During testing we need to find the distance from test vector to all the points in the data to find k nearest neighbours which is computationally expensive. To do this, Even during testing we need to store entire training data, there is no model learning as such here.

Whereas in KMeans we only need to find the distance to centroids to classify the test vector. This is less expensive compared to KNN. During testing we only needs to store centroids.

**Classification:**
KNN outputs most common label of neighbours. It is supervised learning algorithm as we are providing with labels along with training data.

General KMeans model only groups the given data into clusters, also during prediction phase it outputs the group index in which test vector belongs to. But to output a label, we need to manually assign a labels to group/ we have to implement another model for that part.
